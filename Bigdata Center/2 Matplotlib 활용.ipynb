### 라인 플롯

```python
# style.plot ==> 플롯을 그리는 함수
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), "-", linewidth=1, label="sin(x)") # (x,y,line style('-'solid,'--'dotted),line deapth,label)
plt.plot(x, np.cos(x), "--", linewidth=1, label="cos(x)")
plt.xlim(0, 10) # limit ==> 값들의 범위 설정
plt.ylim(-1, 1);
plt.title("Sine & Cosine Curve") # title ==> 제목 지정
plt.xlabel("x") # label ==> 축 이름 설정
plt.legend();# legend ==> 범례를 표시
```

```python
# style.context ==> plot의 스타일을 일시적으로 변경(이 셀에서만)
with plt.style.context("ggplot"):
    plt.plot(x, np.sin(x), '-.', linewidth=1, label="sin(x)")
    plt.plot(x, np.cos(x), ':', linewidth=1, label="cos(x)")
    plt.xlim(0, 10)
    plt.ylim(-1, 1)
    plt.title("Sine & Cosine Curve")
    plt.xlabel("x")
    plt.legend();
```

```python
# Figure(면), Axes(축) 객체 사용 (plt를 대신하여 사용)
fig = plt.figure() # figure 객체 얻기
ax = plt.axes() # axes 객체 얻기

x = np.linspace(0, 10, 100)

ax.plot(x, np.sin(x), "-", linewidth=1, label="sin(x)")
ax.plot(x, np.cos(x), "--", linewidth=1, label="cos(x)")
ax.set_xlim([0, 10]) # set ==> 값을 변경
ax.set_ylim([-1, 1])
ax.set_title("Sine & Cosine Curve")
ax.set_xlabel("x")
ax.legend();
```

```python
# subplots ==> 도면을 분리하여 그림
fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(x, np.sin(x), 'b-', linewidth=0.5, label="sin(x)") # 'b-' blue solid
ax[0].set_title("Sine & Cosine Curve")
ax[0].set(xlim=[0, 10], ylim=[-1, 1]) # set() ==> 여러개의 속성들 한 번에 설정
ax[0].legend()

ax[1].plot(x, np.cos(x), 'g-', linewidth=0.5, label="cos(x)") # 'g-' green solid
ax[1].set(xlim=[0, 10], ylim=[-1, 1], xlabel="x")
ax[1].legend();
```

```python
# pylab 인터페이스를 사용한 매트랩 스타일
plt.subplot(2, 1, 1) # 전체 도면의 형태 설정(2행 1열의 첫 번쨰)
plt.plot(x, np.sin(x), 'b-', linewidth=0.5, label="sin(x)")
plt.title("Sine & Cosine Curve")
plt.axis([0, 10, -1, 1]) # axis ==> 값들의 범위 설정(limit과 유사)
plt.xticks(np.arange(0, 10, 2), []) # 격자?
plt.legend()

plt.subplot(2, 1, 2) # 전체 도면의 형태 설정(2행 1열의 두 번쨰)
plt.plot(x, np.cos(x), 'g-', linewidth=0.5, label="cos(x)")
plt.xlabel("x")
plt.axis([0, 10, -1, 1])
plt.legend();
```

```python
plt.plot(x, np.sin(x), "-", linewidth=1)
plt.title("Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.margins(0, 0)
plt.axis("scaled");
```

```python
# equal ==> 축들이 동일한 크기로 유지
plt.plot(x, np.sin(x), "-", linewidth=1)
plt.title("Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.margins(0, 0)
plt.axis("equal");
```

### 스캐터 플롯

```python
# plot 함수를 이용한 스캐터 플롯(색, 패턴 등을 지정하는 함수가 약함)
np.random.seed(0)

x = np.random.randn(100) # randn ==> n은 normal distribution(평균0,편차1 정규분포)
y = np.random.randn(100)

plt.plot(x, y, "ko", ms=2); # (x,y,color/dot('k'black,'o'circle point),marker size)
```

```python
# scatter ==> 스캐터 플롯을 더 다채롭게 그릴 수 있게 함
np.random.seed(0)

x = np.random.randn(100) # 표준정규분포를 따르는 난수
y = np.random.randn(100)
c = np.random.rand(100) # color 난수 생성
s = np.random.rand(100) * 500 # size 난수 생성

plt.scatter(x, y, c=c, s=s, alpha=0.2, cmap='viridis') # (x,y,color(반드시 수치형),size,opacity,colormap('virdis' 등))
plt.colorbar(); # colorbar ==> 색의 값 의미 표시
```

```python
# seaborn에 있는 iris 데이터
from seaborn import load_dataset
iris = load_dataset("iris")
iris.info()
```

```python
iris.head()
```

```python
import pandas as pd

iris["species_cat"] = pd.Categorical(iris.species) # categorical ==> columns를 범주형 데이터로 변환
iris.info() # species_cat는 category 형태(R에서 factor 형태와 비슷)
```

```python
plt.scatter(iris.sepal_length, iris.sepal_width,
            c=iris.species_cat.cat.codes, alpha=0.2, cmap="viridis") # cat ==> category가 가지고 있는 객체로 수치형 데이터를 가짐 
plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max()) # 최대 최소값을 제한 범위로 설정
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
plt.colorbar(ticks=range(3), label="species")
```

```python
# unique ==> 중복값을 제외한 값
species = iris.species.unique()
species
```

```python
for sp in species:
    x = iris.loc[iris["species"] == sp, "sepal_length"]
    y = iris.loc[iris["species"] == sp, "sepal_width"]
    plt.scatter(x, y, label=sp, alpha=0.2)
plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max())
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
plt.legend();
```

### 히스토그램

```python
np.random.seed(0)

mu = 72
sigma = 8
x = np.random.normal(mu, sigma, 10000) # 평균 72, 편차 8, 데이터 수 10000

plt.hist(x, rwidth=0.9, bins=16); # (x,bar deapth,number of bar)
```

```python
# seaborn에 있는 tips 데이터
tips = load_dataset("tips")
tips.info()
```

```python
plt.hist(x = tips.tip, rwidth=0.9, bins = 20);
plt.title("tips")
plt.xlabel("tip")
plt.ylabel("frequency");
```

```python
# 히스토그램과 확률밀도 함수
import scipy as sp # scipy는 통계와 관련된 함수가 많이 있음

bins = 20
plt.hist(x = tips.tip, rwidth=0.9, bins = bins, density=True); # density=True ==> y축을 밀도로
plt.title("tips")
plt.xlabel("tip")
plt.ylabel("density") # frequency -> density

mu, sigma = tips.tip.mean(), tips.tip.std() # sp에 평균과 밀도를 전해줘야 함
tip_min = tips.tip.min()
tip_max = tips.tip.max()
x = np.linspace(tip_min, tip_max, bins)
y = sp.stats.norm(mu, sigma).pdf(x) # pdf ==> 확률 밀도 함수
plt.plot(x, y);
```

```python
labels = []
tip_list = []

for label, df_per_sex in tips.groupby("sex"):
    labels.append(label) # male과 female
    tip_list.append(df_per_sex["tip"].tolist()) # tolist ==> array를 list로 전환

plt.boxplot(tip_list, labels=labels)
plt.title("tips")
plt.xlabel("sex")
plt.ylabel("tip");
```

### 이미지 플롯

```python
# sklearn.datasetsd의 load_digits에 있는 손글씨 데이터
from sklearn.datasets import load_digits

digits = load_digits()
digits.keys()
```

```python
# 하나의 이미지가 8*8배열로 구성(숫자들은 밝기 정보를 표현 - 0은 흰색)
digits.images[0] # 0은 메모리만 차지하므로 희소행렬로 변환해야함
```

```python
fig, ax = plt.subplots(10, 10, figsize=(5, 5)) # subplots로 10행 10열로 구성
for i, ax_i in enumerate(ax.flatten()): # flatten ==> 다차원 구조를 1차원으로 변경(지금은 2->1)
    ax_i.imshow(digits.images[i], cmap="binary") # imshow ==> 이미지를 그림
    ax_i.set(xticks=[], yticks=[]) # x,y 정보를 기술하지 않고 공간만 구성
```
